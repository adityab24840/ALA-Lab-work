```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
df_google = pd.read_csv("GOOG.csv")
closing_price = df_google["close"].values
window_size = 30
window = np.ones(window_size) / window_size
smoothed_closing_price = np.convolve(closing_price, window, mode="same")

plt.figure(figsize=(16, 8)) 
plt.plot(df_google["date"], closing_price, label="Original") 
plt.plot(df_google["date"], smoothed_closing_price, label="Smoothed") 
plt.xlabel("Date") 
plt.ylabel("Closing Price") 
plt.title("Google Stock Prices") 
plt.legend() 
plt.show()
```
